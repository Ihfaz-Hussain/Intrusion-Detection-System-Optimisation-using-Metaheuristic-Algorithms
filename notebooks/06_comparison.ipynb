```python
# Cell 1 — Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# Cell 2 — Load All Results
base = Path.cwd()
if not (base / "data").exists():
    base = base.parent

baseline = {
    "method": "Baseline RF",
    # paste your actual baseline values here
    "accuracy": 0.7900,
    "precision": 0.9990,
    "recall": 0.6344,
    "f1": 0.7760,
    "fpr": 0.0009,
    "n_features": 78,
    "runtime_s": 0
    
    
}

ga_results  = pd.read_csv(base / "data/processed/ga_results.csv").to_dict("records")[0]
pso_results = pd.read_csv(base / "data/processed/pso_results.csv").to_dict("records")[0]
sa_results  = pd.read_csv(base / "data/processed/sa_results.csv").to_dict("records")[0]

results = pd.DataFrame([baseline, ga_results, pso_results, sa_results])
cols = ["method", "accuracy", "precision", "recall", "f1", "fpr", "n_features", "runtime_s"]
results = results[cols]

print(results.to_string(index=False))

# Cell 3 — Comparison Bar Charts
metrics = ["accuracy", "precision", "recall", "f1", "fpr"]
fig, axes = plt.subplots(1, len(metrics), figsize=(20, 5))

for ax, metric in zip(axes, metrics):
    sns.barplot(data=results, x="method", y=metric, ax=ax, palette="Blues_d")
    ax.set_title(metric.upper().replace("_", " "))
    ax.set_xticklabels(ax.get_xticklabels(), rotation=15)
    ax.set_ylim(0, 1)

plt.suptitle("Metaheuristic vs Baseline — Performance Comparison", fontsize=13)
plt.tight_layout()
plt.savefig("comparison_metrics.png", dpi=150)
plt.show()

# Cell 4 — Trade-off: F1 vs Number of Features
plt.figure(figsize=(8, 5))
for _, row in results.iterrows():
    plt.scatter(row["n_features"], row["f1"], s=150, label=row["method"])
    plt.annotate(row["method"], (row["n_features"], row["f1"]),
                 textcoords="offset points", xytext=(5, 5))

plt.xlabel("Number of Features")
plt.ylabel("F1 Score")
plt.title("Trade-off: F1 Score vs Feature Count")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("tradeoff_f1_features.png", dpi=150)
plt.show()

# Cell 5 — Summary Table (for your paper)
print("\nFinal Summary Table:")
print(results.to_string(index=False))
results.to_csv(base / "data/processed/final_comparison.csv", index=False)
print("Saved to final_comparison.csv")
```
